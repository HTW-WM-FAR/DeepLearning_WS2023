```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('sep100.csv')
```

```python
df["Date"] = pd.to_datetime(df.Date)
df = df.set_index("Date")
df
```

```python
columns_list = sorted(df)
with open('column_names', 'w') as fp:
    fp.write("\n".join(str(item) for item in columns_list))
```

```python
def get_top_10_companies(date_row):
    top_10 = date_row.nlargest(10)
    return top_10
```

```python
top_10 = df.apply(get_top_10_companies, axis=1)
top_10
```

## Top 10 Stocks

```python
most_recent_date = top_10.index[-1]

with open('top_10.txt', 'w') as fp:
    fp.write(f"\nDate: {most_recent_date}\n")
    for company, value in top_10.loc[most_recent_date].items():
        fp.write(f"{company}: {value}\n")
    fp.write("\n")
```

```python
top10_df = df[["ADBE", "AVGO", "BKNG", "BLK", "COST", "LMT", "NFLX", "NVDA", "TMO", "UNH"]]
top10_df
```

### Finding Co-relation with Amazon (as an example)

```python
def find_correlated_feature(company, target_feature, threshold=0.9):
    corr_matrix = top10_df.corr()
    highly_correlated_features = corr_matrix[abs(corr_matrix[target_feature]) >= threshold].index.tolist()
    highly_correlated_features.remove(target_feature)
    return highly_correlated_features

avgo_corr = find_correlated_feature(top10_df, "AVGO", threshold = 0.8)
```

```python
plt.figure(figsize = (14,8))
sns.heatmap(df[avgo_corr].corr(), annot = True, cmap = "viridis")
plt.show()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

fig = make_subplots(rows=4, cols=1, shared_xaxes=True, vertical_spacing=0.05)

scatter1 = go.Scatter(x=top_df["ADBE"], y=top_df["AMZN"], mode="markers", name="ADBE vs AMZN")
fig.add_trace(scatter1, row=1, col=1)

scatter2 = go.Scatter(x=top_df["GOOG"], y=top_df["AMZN"], mode="markers", name="GOOG vs AMZN")
fig.add_trace(scatter2, row=2, col=1)

scatter3 = go.Scatter(x=top_df["NFLX"], y=top_df["AMZN"], mode="markers", name="NFLX vs AMZN")
fig.add_trace(scatter3, row=3, col=1)

scatter4 = go.Scatter(x=top_df["V"], y=top_df["AMZN"], mode="markers", name="V vs AMZN")
fig.add_trace(scatter4, row=4, col=1)

fig.update_layout(
    height=800,
    title_text="Correlation between Amazon, Adobe, Google, Netflix & Visa",
)

fig.show()
```

### Moving Average for Amazon, Google and Visa

```python
top_df["AMZN_10"]= top_df["AMZN"].rolling(10).mean()
top_df["AMZN_30"]= top_df["AMZN"].rolling(30).mean()
top_df["AMZN_50"]= top_df["AMZN"].rolling(50).mean()

top_df["GOOG_10"]= top_df["GOOG"].rolling(10).mean()
top_df["GOOG_30"]= top_df["GOOG"].rolling(30).mean()
top_df["GOOG_50"]= top_df["GOOG"].rolling(50).mean()

top_df["V_10"]= top_df["V"].rolling(10).mean()
top_df["V_30"]= top_df["V"].rolling(30).mean()
top_df["V_50"]= top_df["V"].rolling(50).mean()
```

```python
def plot_moviang_avg(df):
    top_df = df.iloc[5200:].copy()
    fig, ax = plt.subplots(2, 2, figsize=(15, 10))
    line_width = 1.5

    ax[0, 0].plot(top_df[["AMZN", "AMZN_10", "AMZN_30", "AMZN_50"]], linewidth=line_width)
    ax[0, 0].set_title("Amazon")
    ax[0, 0].legend(["AMZN", "AMZN_10", "AMZN_30", "AMZN_50"])  
    ax[0, 1].plot(top_df[["V", "V_10", "V_30", "V_50"]], linewidth=line_width)
    ax[0, 1].set_title("Visa")
    ax[0, 1].legend(["V", "V_10", "V_30", "V_50"])  
    ax[1, 0].plot(top_df[["GOOG", "GOOG_10", "GOOG_30", "GOOG_50"]], linewidth=line_width)
    ax[1, 0].set_title("Google")
    ax[1, 0].legend(["GOOG", "GOOG_10", "GOOG_30", "GOOG_50"]) 

    plt.tight_layout()  
    plt.show()

plot_moviang_avg(top_df)
```
